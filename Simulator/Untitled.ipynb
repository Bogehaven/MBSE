```python
import plotly.express as px
import pandas as pd
import glob
```

```python
def load_files(file_list):
    df_hist = pd.DataFrame()
    df_samp = pd.DataFrame()

    for f in file_list:
        if len(f.split("/")[-1].split('.')[0].split('_')) > 3:
            continue
        _, typ, cat = f.split("/")[-1].split('.')[0].split('_')
        df = pd.read_csv(f)
        df.columns.values[0] = cat

        if typ == "history":
            df_hist = pd.concat([df_hist, df], axis=1)
        elif typ == "sampled":
            df_samp = pd.concat([df_samp, df], axis=1)
            
    return df_hist, df_samp
```

```python
file_list = (glob.glob("results/results/*"))
```

```python
df_hist, df_samp = load_files(file_list)
```
